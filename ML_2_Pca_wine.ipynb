```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Wine Dataset.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
x = data.drop('class',axis = 1)
```

```python
y = data['class']
```

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
np.float = float
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x_scaled)
```

```python
principal_components
```

```python
data = pd.DataFrame(principal_components)
```

```python
data.head()
```

```python
data['class']=y
```

```python
data.head()
```

```python
data.columns=["PC1","PC2","class"]
data.head()
```

```python
data
```

```python
markers = ['s', 'x','o']
colors = ['r', 'b','g']
sns.lmplot(x="PC1", y="PC2", data=data, hue='class', markers=markers,fit_reg=False,legend=True)
```
